```python
class Producto:
    def __init__(self, nombre, precio):
        self.nombre = nombre
        self.precio = precio

    def __str__(self):
        return f"{self.nombre}: ${self.precio}"


class Combo:
    def __init__(self, nombre, precio, descripcion):
        self.nombre = nombre
        self.precio = precio
        self.descripcion = descripcion

    def __str__(self):
        return f"{self.nombre} ({self.descripcion}): ${self.precio}"


class Cine:
    def __init__(self):
        self.combos = [
            Combo("Personal", 23000, "Crispetas + Gaseosa"),
            Combo("Compartir", 40000, "Crispetas + Gaseosa + Nachos"),
            Combo("Familiar", 50000, "Crispetas + Gaseosa + Perro + Nachos"),
        ]

        self.productos_individuales = [
            Producto("Crispetas", 20000),
            Producto("Gaseosa", 6000),
            Producto("Nachos", 18000),
            Producto("Perro", 12000),
            Producto("Chocolate", 5000)
        ]
        self.total = 0
        self.descuento_tarjeta = 0.15

    def mostrar_menu(self):
        print("---- Menú de Combos ----")
        for combo in self.combos:
            print(combo)

        print("\n---- Productos Individuales ----")
        for producto in self.productos_individuales:
            print(producto)

    def seleccionar_combo(self, nombre_combo, cantidad):
        for combo in self.combos:
            if combo.nombre == nombre_combo:
                subtotal = combo.precio * cantidad
                self.total = self.total + subtotal
                print(f"Has agregado {cantidad} x {combo.nombre} - Subtotal: ${subtotal:.2f}")
                return
        print("Combo no válido. Inténtelo de nuevo.")

    def seleccionar_producto(self, nombre_producto, cantidad):
        for producto in self.productos_individuales:
            if producto.nombre == nombre_producto:
                subtotal = producto.precio * cantidad
                self.total = self.total + subtotal
                print(f"Has agregado {cantidad} x {producto.nombre} - Subtotal: ${subtotal:.2f}")
                return
        print("Producto no válido. Inténtelo de nuevo.")

    def elegir_metodo_pago(self):
        print("\nMétodos de pago disponibles: tarjeta, efectivo, PSE, transferencia")
        metodo = input("Seleccione un método de pago: ").lower()
        if metodo in ["tarjeta", "efectivo", "pse", "transferencia"]:
            print(f"Ha seleccionado pagar con {metodo}.")
        else:
            print("Método de pago no válido. Inténtelo de nuevo.")
            self.elegir_metodo_pago()

    def aplicar_descuento(self):
        tarjeta_cine = input("¿Tiene tarjeta del cine? (si/no): ").lower()
        if tarjeta_cine == "si":
            descuento = self.total * self.descuento_tarjeta
            self.total -= descuento
            print(f"Se ha aplicado un descuento del 15%. Descuento: ${descuento:.2f}")
        elif tarjeta_cine == "no":
            print("No se aplicó descuento.")
        else:
            print("Opción no válida. Inténtelo de nuevo.")
            self.aplicar_descuento()

    def realizar_compra(self):
        print("\nBienvenido al Cine, ¿Qué desea comprar?\n")
        self.mostrar_menu()

        while True:
            opcion = input("\nEscriba 'combo' para elegir un combo, 'producto' para un producto individual, o 'salir' para finalizar: ").lower()

            if opcion == "combo":
                combo = input("Seleccione un combo (Combo personal, Combo compartir, Combo familiar): ").capitalize()
                cantidad = int(input(f"Ingrese la cantidad de {combo} que desea comprar: "))
                self.seleccionar_combo(combo,cantidad)

            elif opcion == "producto":
                producto = input("Seleccione un producto (Crispetas, Gaseosa, Nachos, Perro, Chocolate): ").capitalize()
                cantidad = int(input(f"Ingrese la cantidad de {producto} que desea comprar: "))
                self.seleccionar_producto(producto, cantidad)

            elif opcion == "salir":
                break
            else:
                print("Opción no válida. Inténtelo de nuevo.")

        print(f"\nSubtotal: ${self.total:.2f}")
        self.aplicar_descuento()
        print(f"Total a pagar después de descuentos: ${self.total:.2f}")
        self.elegir_metodo_pago()
        print("Gracias por su compra. ¡Disfrute la película!")


cine = Cine()
cine.realizar_compra()

```
