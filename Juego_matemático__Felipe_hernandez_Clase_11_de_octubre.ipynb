#Ejercicio
juego de matemáticas donde el usuario elija entre suma, resta, multiplicación, división y preguntas mezcladas de esas operaciones. Al usuario le toca responder bien 10 preguntas seguidas para ganar y que se acabe el programa.
Realizarlo dos veces, una con funciones y luego con funciones anónimas

```python
#Funciones normales
import random

def operar(a, b, op):
    if op == 1:
        return a + b
    elif op == 2:
        return a - b
    elif op == 3:
        return a * b
    elif op == 4:
        return round(a / b, 1) if b != 0 else 'undefined'

def jugar(op):
    correctas = 0
    while correctas < 10:
        a, b = random.randint(1, 10), random.randint(1, 10)
        operacion = op if op != 5 else random.randint(1, 4)
        simbolo = {1: '+', 2: '-', 3: '*', 4: '/'}[operacion]
        respuesta = operar(a, b, operacion)

        # Si es una división, avisar sobre el decimal
        if operacion == 4:
            print(f"{a} {simbolo} {b} (Escribe el resultado con un solo decimal)")
        else:
            print(f"{a} {simbolo} {b}")

        if input("Respuesta: ") == str(respuesta):
            correctas += 1
            print(f"¡Correcto! Llevas {correctas}/10")
        else:
            print(f"Incorrecto. La respuesta correcta era {respuesta}. ¡Inténtalo de nuevo!")
            correctas = 0
    print("¡Ganaste!")

def menu():
    print("Elige una operación:\n1. Suma\n2. Resta\n3. Multiplicación\n4. División\n5. Mezcla")
    op = int(input("Elige una opción (1-5): "))
    jugar(op)

menu()
```

```python
#Funciones anónimas
import random

def jugar(op):
    correctas = 0
    while correctas < 10:
        a, b = random.randint(1, 10), random.randint(1, 10)

        # Definir operaciones usando lambdas sin diccionario
        if op == 1:  # Suma
            operacion = lambda a, b: a + b
            simbolo = '+'
        elif op == 2:  # Resta
            operacion = lambda a, b: a - b
            simbolo = '-'
        elif op == 3:  # Multiplicación
            operacion = lambda a, b: a * b
            simbolo = '*'
        elif op == 4:  # División
            operacion = lambda a, b: round(a / b, 1) if b != 0 else 'undefined'
            simbolo = '/'
        else:  # Mezcla aleatoria
            op = random.randint(1, 4)
            continue  # Volver al ciclo para generar una nueva operación aleatoria

        respuesta = operacion(a, b)

        if op == 4 and a % b != 0:  # División no exacta
            print(f"{a} {simbolo} {b} (responde con un decimal)")
        else:
            print(f"{a} {simbolo} {b}")

        if input("Respuesta: ") == str(respuesta):
            correctas += 1
            print(f"¡Correcto! Llevas {correctas}/10")
        else:
            print(f"Incorrecto. La respuesta correcta era {respuesta}. ¡Inténtalo de nuevo!")
            correctas = 0
    print("¡Ganaste!")

def menu():
    print("Elige una operación:\n1. Suma\n2. Resta\n3. Multiplicación\n4. División\n5. Mezcla")
    op = int(input("Elige una opción (1-5): "))
    jugar(op)

menu()
```
